```python
import numpy as np
import time

n = 10000
x = np.random.randn(n,n)

def matrix_mult(A, B):
    """Multiplies two matrices A and B."""
    if len(A[0]) != len(B):
        raise ValueError("Number of columns in A must be equal to number of rows in B")

    result = []
    for i in range(len(A)):
        row = []
        for j in range(len(B[0])):
            sum = 0
            for k in range(len(B)):
                sum += A[i][k] * B[k][j]
            row.append(sum)
        result.append(row)
    return result
```
